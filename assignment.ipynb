```python
from MyRE.regex_to_NFA import RegexToNFA , WriteJsonFile
from MyRE.NFA_to_DFA import NFA_to_DFA , minimize_DFA
from MyRE.visualizer import draw_DFA, draw_NFA

google_drive = False

if google_drive:
    path = '/content/gdrive/MyDrive/compilers_assignment/'
else:
    path = ''
```

```python
# regex = input("Enter a regex: ")
regex = "((a(b*))|((c?)+)|([A-Z123])|(ex))"
# regex = "(((ab)c)d*)"
NFA = RegexToNFA(regex)

NFA = WriteJsonFile(NFA, path+"Json/NFA.json")
draw_NFA(NFA , path+"output/NFA")
```

```python
DFA = NFA_to_DFA(NFA)
minimized_DFA = minimize_DFA(DFA , "Json/DFA.json")
draw_DFA(minimized_DFA, "output/DFA")
```
